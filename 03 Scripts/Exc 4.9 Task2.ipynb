The histogram of the "order_hour_of_day" column shows that most orders are placed between 10 AM and 4 PM, peaking around 12 PM to 1 PM.

# Creating line chart

# sampling

```python
dev = np.random.rand(len(df)) <= 0.7
```

```python
# store 70% of the sample in the dataframe big
big = df[dev]
```

```python
# store 30% of the sample in the dataframe small
small = df[~dev]
```

```python
len(df)
```

```python
len(big)+len(small)
```

#  Reducing samples to necessary columns

```python
df_1=small[['order_hour_of_day','prices']]
```

# Create line chart  whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day

```python
line = sns.lineplot(data = df_1, x = 'order_hour_of_day',y = 'prices')
```

The line chart indicates significant spikes in prices during the early morning hours (0-5 AM), especially around 1 AM and 4 AM. From 6 AM to 8 PM, prices stabilize between 10 and 15, with minor fluctuations. This suggests higher variability in expenditures during early hours and consistent pricing throughout the day and evening.
